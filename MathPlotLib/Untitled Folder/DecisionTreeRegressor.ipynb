```python
from sklearn.datasets import load_iris
import pandas as pd
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import tree
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris_data=load_iris(as_frame=True)
```

```python
df=pd.DataFrame(iris_data.data)
df.head()
```

```python
iris_data.data.corr()
```

```python
petal_length=df["sepal length (cm)"]
petal_width=df["petal width (cm)"]
```

```python
classes=iris_data["target"]
```

```python
plt.scatter(petal_length,petal_width,c=classes)
```

```python
x=petal_length.values
y=petal_width.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,shuffle=True,random_state=14)
```

```python
reg=tree.DecisionTreeRegressor()
```

```python
history=reg.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
y_pred=reg.predict(x_test.reshape(-1,1))
```

```python
history.score(x_test.reshape(-1,1),y_test)
```

```python
my_x=[[5]]
my_pred=reg.predict(my_x)
my_pred
```

```python
a ,b =np.polyfit(x_train,y_train,1)#y=ax+b
```

```python
plt.scatter(x_test,y_test)
plt.scatter(my_x,my_pred)
plt.plot(x,a*x+b)
```
